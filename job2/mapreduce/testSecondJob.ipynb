```python
import pandas as pd
df1 = pd.read_csv('../../dataset/historical_stock_prices.csv')
df2 = pd.read_csv('../../dataset/historical_stocks.csv')
df1.head()
```

### Read both datasets

```python
df2.head()
```

```python
df1.shape
```

```python
df2.shape
```

### Merge the datasets

```python
df = pd.merge(df1, df2, on = 'ticker')
```

```python
df.shape
```

```python
df.head()
```

### Select rows which sectors are basic industries

```python
basic_industries = df[df['sector'] == 'BASIC INDUSTRIES']
basic_industries = basic_industries.sort_values('date')
```

```python
basic_industries.head(10)
```

```python
basic_industries.shape
```

```python
basic_industries.describe()
```

```python
len(basic_industries['ticker'].unique())
```

### select basic industries record in year 2018

```python
basic_industries2018 = basic_industries[basic_industries['date'].str.startswith('2018')]
```

```python
basic_industries2018.shape
```

```python
len(basic_industries2018['ticker'].unique())
```

### Add year column

```python
pd.options.mode.chained_assignment = None
basic_industries2018['year'] = basic_industries2018.date.str[0:4]
```

```python
basic_industries2018.head()
```

```python
basic_industries2018.shape
```

### Group by ticker and year and get first date

```python
basic_industries2018GroupedFirstDate = basic_industries2018.groupby(['ticker', 'year']).first()
basic_industries2018GroupedFirstDate.head()
```

```python
basic_industries2018GroupedFirstDate.shape
```

```python
startingClosePrice = basic_industries2018GroupedFirstDate['close'].sum()
startingClosePrice
```

### Group by ticker and year and get last date

```python
basic_industries2018GroupedLastDate = basic_industries2018.groupby(['ticker', 'year']).last()
basic_industries2018GroupedLastDate.head()
```

```python
basic_industries2018GroupedLastDate.shape
```

```python
finalClosePrice = basic_industries2018GroupedLastDate['close'].sum()
finalClosePrice
```

```python
percentDifference = (finalClosePrice - startingClosePrice)/startingClosePrice
percentDifference
```
